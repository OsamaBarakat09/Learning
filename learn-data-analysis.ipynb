```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns',10)
pd.set_option('display.max_rows',10)

meta = pd.read_csv("../input/results-of-companies-stocks-for-one-year/META.csv")
appl = pd.read_csv('../input/results-of-companies-stocks-for-one-year/AAPL.csv')
amc = pd.read_csv('../input/results-of-companies-stocks-for-one-year/AMC.csv')
cdxs = pd.read_csv('../input/results-of-companies-stocks-for-one-year/CDXS.csv')
amzn = pd.read_csv('../input/results-of-companies-stocks-for-one-year/AMZN.csv')
class Meta(object):
    def __init__(self, shape, info, describe, value, column, index):
        self.shape= shape
        self.info= info
        self.describe = describe
        self.value = value
        self.column = column
        self.index = index
    def Exploring(self):
        print(self.shape) # contains a tuple that holds the number of rows followed by the number of columns
        print(self.info) # The info method displays the names of columns, the data types they contain, and whether they have any missing values.
        print(self.describe) # quick overview of numeric variables
    def Components(self):
        print(self.value) # contains the data values in a 2-dimensional NumPy array.
        print(self.column) # columns name
        print(self.index) # row number or names

res_Meta = Meta(meta.shape,'''meta.info()''', meta.describe(),
                meta.values, meta.columns, meta.index)



class Sorting_Subsetting_Meta():
    def __init__(self, sort, subset, isin):
        self.sort =sort 
        self.subset = subset
        self.isin = isin
    def Sorting(self):
        print(self.sort) # select particular columns and sort data fram by it
    def Subsetting(self):
        print(self.subset)
    def Isin(self):
        print(self.isin)

res_Sorting_Subsetting_Meta = Sorting_Subsetting_Meta(meta.sort_values('Open',ascending=True),
                                           meta.Open,
                                           meta[meta['Date'].isin(['2022-08-01'])]) 


class Sunmmerizing_Statistics():
    def __init__(self, mean, median,mode, Max, Min,var, std, Sum, quantile,Agg, cumsum, cummax,cummin):
        self.mean = mean
        self.median = median
        self.mode= mode
        self.Max = Max
        self.Min = Min
        self.var = var
        self.std = std
        self.Sum = Sum
        self.quantile = quantile 
        self.Agg = Agg
        self.cumsum = cumsum
        self.cummax = cummax
        self.cummin = cummin

    def Display(self):
        print(self.mean) # The center of the data
        print(self.median)
        print(self.mode)
        print(self.Max)
        print(self.Min)
        print(self.var)
        print(self.std)
        print(self.Sum)
        print(self.quantile)
        print(self.Agg) # Compute custom summary statistics
    def Cumulative(self):
        print(self.cumsum)
        print(self.cummax)
        print(self.cummin)
res_Sunmmerizing_Statistics = Sunmmerizing_Statistics(meta.Open.mean(),
                                                      meta.Open.median(),
                                                      meta.Open.mode(),
                                                      meta.Open.max(),
                                                      meta.Open.min(),
                                                      meta.Open.var(),
                                                      meta.Open.std(),
                                                      meta.Open.sum(),
                                                      meta.Open.quantile(),
                                                      meta.Open.agg([max,min,sum]),
                                                      meta.Open.cumsum(),
                                                      meta.Open.cummax(),
                                                      meta.Open.cummin())

class Counting():
    def __init__(self,drop_duplicate, value_counts, pivot, pivot_work,axis):
        self.drop_duplicate = drop_duplicate
        self.value_counts = value_counts
        self.pivot = pivot
        self.pivot_work = pivot_work
        self.axis = axis
    def Display(self):
        print(self.drop_duplicate) # Removing rows the contain values that was been listed
        print(self.value_counts) # use to count number of each value or rows
    def Pivot(self):
        print(self.pivot)
        print(self.pivot_work)
    def Axis(self):
        print(self.axis)
        
        
res_Counting = Counting(meta.drop_duplicates(subset=["Open",'Close']),
               meta.value_counts(sort=True, ascending=True),
               meta.pivot_table(values='Close', index='Open', columns='Date',
                                aggfunc=np.median, fill_value=0, margins=True),
                meta.pivot_table(values='Close', index='Open', columns='Date'),
                       meta.mean(axis='index'))
                        # aggfunc: using when the row contain multiplue values 

class Explicit_index():
    def __init__(self,Set, reset,Loc, sort,inner, rowAndcolumn):
        self.set = Set
        self.reset = reset
        self.Loc = Loc
        self.sort = sort
        # self.outer = outer
        self.inner=inner
        self.rowAndcolumn = rowAndcolumn
    def Display(self): 
        print(self.set) # putting any column as index
        self.Reset_index()
        self.Subsettin_On_Duplicated()
        
    def Reset_index(self):
        print(self.reset) # Remove index and replace it by sorting number from 1,2,....
   
    def Subsettin_On_Duplicated(self):
        print(self.Loc)
    def Sort_Index(self):
        print(self.sort)
    
    def Slicing_Inner_Index(self):
        print(self.inner)
    #def Slicing_Outer_Index(self):
       # print(self.outer)      meta.set_index(['Date','Volume']).loc[('2021-08-04',"14180600"):('2021-08-06',"8925000")],
    def Subsetting_By_RowAndColumn(self):
        print(self.rowAndcolumn)

res_Explicit_index = Explicit_index(meta.set_index('Open'),
                                    meta.reset_index(drop=False),
                                    meta.set_index('Date').loc[['2021-08-04','2022-07-27']],
                                    meta.sort_index(level=["Date","Open"], ascending=[True]),
                                    meta.set_index('Date').loc['2021-08-04':'2022-07-27'],
                                    meta.iloc[1:10, 1:5 ])


class Visualizing_Data():
    def __init__(self,hist, plot):
        self.hist = hist
        self.plot = plot
    def Hist(self):
        print(self.hist)
    def Plot(self):
        print(self.plot)
res_Visualizing_Data = Visualizing_Data(meta.Open.hist(bins=50),
                                        meta.plot(kind='line',#kind = lina, bar, scatter
                                                  x='Date',y='Close',title="Distribution of Met Stock",rot=90))


print(amzn)






```
